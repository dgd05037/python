OpenCV 이미지 합치기


```python
import cv2
import matplotlib.pyplot as plt

image1 = cv2.imread('cat.jpg')
image2 = cv2.imread('cat1.jpg')

print(image1.size)
print(image1.shape)
print(image2.size)
print(image2.shape)

expand = cv2.resize(image2, None, fx=275/204,fy=1, interpolation = cv2.INTER_CUBIC)
print(expand.shape)
plt.imshow(cv2.cvtColor(image2,cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(expand,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.full((512, 512, 3),255, np.uint8)
image = cv2.line(image,(0,0),(255,255),(255,0,0),3)

plt.imshow(image)
plt.show()
```

```python
image1 = cv2.imread('cat.jpg')
image_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(image_gray, 127, 255, 0)

plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_GRAY2RGB))
plt.show()

contours = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)[1]
image = cv2.drawContours(image, contours, -1, (0, 255, 0), 4)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
