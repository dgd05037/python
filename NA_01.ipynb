```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_basic = cv2.imread('image.jpg', cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
plt.imsave('result1.png',img_basic)
```

```python
image = img_basic
image[0:100,0:100] = [0,0,0]
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
print(image.size)
print(image.shape)
```

```python
roi = image[1000:2000, 400:1000]
image[2000:3000, 1000:1600] = roi
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image1 = cv2.imread('image.jpg')
image[:,:,2] = 0

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
expand = cv2.resize(image1, None, fx=2.0, fy=2.0, interpolation = cv2.INTER_CUBIC)
plt.imshow(cv2.cvtColor(expand, cv2.COLOR_BGR2RGB))
plt.show()

shrink = cv2.resize(image1, None, fx=0.5, fy=0.5, interpolation = cv2.INTER_AREA)
plt.imshow(cv2.cvtColor(shrink, cv2.COLOR_BGR2RGB))
plt.show()
```
